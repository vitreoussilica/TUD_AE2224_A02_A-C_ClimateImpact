```python
import cartopy.crs as ccrs
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10,5), dpi=150)
#ax = plt.axes(projection=ccrs.Orthographic(central_longitude=46, central_latitude=19))
ax = plt.axes(projection=ccrs.PlateCarree())
ax.stock_img()

extent = [-25, 60, 20, 70]
ax.set_extent(extent)
ax.gridlines()
ax.coastlines(resolution='50m')

for i in range(100):
    lon = acdata.variables['routes_out'].isel(time=0, AirTraf_routes_out=i, AirTraf_properties=0).data
    lat = acdata.variables['routes_out'].isel(time=0, AirTraf_routes_out=i, AirTraf_properties=1).data
    # plt.scatter(lon, lat, transform=ccrs.Geodetic(), marker='.', s=1)
    plt.plot(lon, lat, transform=ccrs.Geodetic())
```
