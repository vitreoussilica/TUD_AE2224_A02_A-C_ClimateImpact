```python
# rain rate 

rsfl = data_ECHAM['rsfl_2d'].isel(time = 0)
rsfc = data_ECHAM['rsfc_2d'].isel(time = 0)

rainr = rsfc + rsfl
rainrateplot = rainr.sel(lat=slice(80., 0.), lon=slice(0, 70)).plot()
```

```python
#snow rate

ssfl = data_ECHAM['ssfl_2d'].isel(time = 0)
ssfc = data_ECHAM['ssfc_2d'].isel(time = 0)
snowr = ssfc + ssfl

snowrateplot = snowr.sel(lat=slice(80., 0.), lon=slice(0, 70)).plot()
```

```python
# relative humidity
humidity = data_ECHAM['rhum'].isel(time = 0, lev = 10)
humidity_plot = humidity.sel(lat=slice(80., 0.), lon=slice(0, 70)).plot()
```

```python
#cloud water
cloudwater = data_ECHAM['xlm1'].isel(time = 0, lev = 10)
cloudwater_plot = cloudwater.sel(lat=slice(80., 0.), lon=slice(0, 70)).plot()
```

```python
#cloud ice

cloudice = data_ECHAM['xim1'].isel(time = 0, lev = 10)
cloudice_plot = cloudice.sel(lat=slice(80., 0.), lon=slice(0, 70)).plot()
```

```python
# vertical winds

vertwinds = data_ECHAM['etadot'].isel(time = 0, ilev = 10)
vertwinds_plot = vertwinds.sel(lat=slice(80., 0.), lon=slice(0, 70)).plot()
```
