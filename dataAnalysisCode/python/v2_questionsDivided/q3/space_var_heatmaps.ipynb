```python
# %matplotlib notebook
import numpy as np
import pandas as pd
import xarray as xr
import cartopy.crs as ccrs
import matplotlib.pyplot as plt
# import ipywidgets as widgets
```

```python
gpdata = xr.open_dataset("sample_data/f00____________20180101_0000_airtraf_gp.nc")
gpdata
```

```python
gpdata.coords['lon'].values
```

```python
lons, lats = np.meshgrid(gpdata.coords['lon'].values, gpdata.coords['lat'].values)
```

```python
plt.figure(figsize=(10,5), dpi=150)
#ax = plt.axes(projection=ccrs.Orthographic(central_longitude=46, central_latitude=19))
ax = plt.axes(projection=ccrs.PlateCarree())
ax.stock_img()

extent = [-25, 60, 20, 70]
ax.set_extent(extent)
ax.gridlines()
ax.coastlines(resolution='50m')

plt.scatter(lons, lats, transform=ccrs.Geodetic())
```

```python
echam = xr.open_dataset('sample_data/f100___________20171202_0000_ECHAM5.nc')
echam
```

```python
echam.variables['pressi'].isel(time = 0, ilev = 10).values
```

```python
press = echam["pressi"].isel(time = 0, ilev = 10)
```

```python
echam["pressi"].isel(time = 0, ilev = 10).sel(lat=slice(80., 0.), lon=slice(0, 70)).plot()
```

```python
p = press.plot(subplot_kws=dict(projection=ccrs.Orthographic(10, 35), facecolor="gray"),
    transform=ccrs.PlateCarree(),)

p.axes.set_global()

p.axes.coastlines()
```

```python
gpnew = xr.open_dataset("sample_data/f100___________20180601_0000_airtraf_gp.nc")
gpnew
```

```python
gpnew["NOx_Emissions_ave"].sel(lev=25, method="nearest").isel(time = 5).values.nonzero()
```

```python
for i in gpnew:
    nz = gpnew[i].values.nonzero()
    print(i)
    print(nz)
```

```python
var = gpnew["ATR20CO2_air_ave"].sel(lev=10, method="nearest").isel(time = 0)
var = var.where(var>1e-20)
ax = plt.axes(projection=ccrs.PlateCarree())
# ax.stock_img()
ax.coastlines()

extent = [-25, 60, 20, 70]
ax.set_extent(extent)

p = var.plot(transform=ccrs.PlateCarree(),)
```
